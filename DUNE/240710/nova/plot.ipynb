```python
def separate_e_p(file):
    e = []
    p = []
    for line in file:
        line = line.replace("\n", "")
        line = line.split(",")
        e.append(float(line[0]))
        p.append(float(line[1]))
    file.close()
    return e, p
```

```python
e1, p1 = separate_e_p(file1)
e2, p2 = separate_e_p(file2)
e3, p3 = separate_e_p(file3)
e4, p4 = separate_e_p(file4)
```
