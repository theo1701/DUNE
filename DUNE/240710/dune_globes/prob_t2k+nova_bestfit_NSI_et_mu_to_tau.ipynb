```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def separate_e_p(filename):
    e = []
    p = []
    with open(filename, "r") as file:
        for line in file:
            line = line.replace("\n", "")
            line = line.split(" ")
            e.append(float(line[0]))
            p.append(float(line[1]))
        file.close()
    return e, p
```
