```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
in1 = open("prob_bestfit_NH_DUNE_nsi_ee+90.dat", "r")
in2 = open("prob_bestfit_NH_DUNE_nsi_ee-90.dat", "r")
in3 = open("prob_bestfit_IH_DUNE_nsi_ee+90.dat", "r")
in4 = open("prob_bestfit_IH_DUNE_nsi_ee-90.dat", "r")
```

```python
def separate_e_p(file):
    e = []
    p = []
    for line in file:
        line = line.replace("\n", "")
        line = line.split(" ")
        if len(line) == 1:
            continue
        e.append(float(line[0]))
        p.append(float(line[1]))
    file.close()
    return e, p
```

```python
e1, p1 = separate_e_p(in1)
e2, p2 = separate_e_p(in2)
e3, p3 = separate_e_p(in3)
e4, p4 = separate_e_p(in4)
```

```python
plt.fill_between(e1, p1, p2, alpha=0.3, edgecolor='red', facecolor='red', label="NH")
plt.fill_between(e3, p3, p4, alpha=0.3, edgecolor='blue', facecolor='blue', label="IH")
plt.ylim(0, 0.175)
plt.xlabel("E[GeV]")
plt.ylabel(r"$P_{\mu e}$")
plt.text(6, 0.1, r"NSI($\epsilon_{ee}=0.7;\delta\in[-\pi/2, \pi/2]$)")
plt.legend()
plt.savefig("prob_bestfit_DUNE_nsi_ee.pdf")
```
